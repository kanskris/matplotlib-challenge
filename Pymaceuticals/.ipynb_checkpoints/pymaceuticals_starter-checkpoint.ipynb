```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_dd = pd.read_csv(mouse_drug_data_to_load)
clinical_td = pd.read_csv(clinical_trial_data_to_load)
```

```python
mouse_dd.head()
```

```python
clinical_td.head()
```

```python
# Combine the data into a single dataset
merged_ds = pd.merge(clinical_td,mouse_dd,how='outer')
```

```python
# Display the data table for preview
merged_ds.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
merged_ds_temp=merged_ds.drop(['Metastatic Sites'],axis=1)
ds_group_dt = merged_ds_temp.groupby(['Drug','Timepoint'])
# Convert to DataFrame
mean_group = ds_group_dt.mean()
mean_df=pd.DataFrame(mean_group).reset_index()
# Preview DataFrame
mean_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
merged_ds_temp1=merged_ds.drop(['Metastatic Sites','Mouse ID'],axis=1)
ds_group_dt = merged_ds_temp1.groupby(['Drug','Timepoint'])
# Convert to DataFrame
sem_group = ds_group_dt.sem()
sem_df=pd.DataFrame(sem_group).reset_index()
# Preview DataFrame
sem_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_df_new=mean_df.pivot(columns='Drug',index='Timepoint')
# Preview that Reformatting worked
mean_df_new.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
sem_df_new=sem_df.pivot(columns='Drug',index='Timepoint')
# Preview that Reformatting worked
sem_df_new.head()
```

```python
# Generate the Plot (with Error Bars)
fig, ax = plt.subplots()
temp=sem_df_new.reset_index()
x_axis=list(temp['Timepoint'])
x_axis
ax.errorbar(x_axis, mean_df_new, sem_df_new, fmt="o")

#ax.set_xlim(-1, len(samples) + 1)

#ax.set_xlabel("Sample Number")
#ax.set_ylabel("Proportion of People Voting Republican")
# Save the Figure

```
